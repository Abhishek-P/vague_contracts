```python
import openai
import os
import pandas
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
prompts = pandas.read_csv("../../experiments/1_falseconsensus/stimuli.csv")
prompts.rename(columns={'Title ':'title'}, inplace=True)
prompts["prompt"] = prompts.header + "\n\n" + prompts.continuation
prompts["question"] = prompts["item"].apply(lambda x: "Using just a number between 0 and 100, estimate the percentage of US English speakers who would believe that the damage that occured is covered under " + x + " as it appears in the policy. ESTIMATE:")
prompts["prompt"] = prompts["prompt"] + prompts["question"]
prompts = prompts[["item","title","version","prompt"]]
prompts.head(1)
```

```python
sampleResponse = openai.Completion.create(model="text-davinci-003", 
                                    prompt=prompts["prompt"][0], 
                                    temperature=0, max_tokens=10)
```

```python
sampleResponse.choices[0].text
# sampleResponse
```

```python
# to make a smaller dataset...
untunedGPT3Predictions = prompts.head(10)
untunedGPT3Predictions
```

```python
# CAUTION: running this costs money! (Or free credits!)
untunedGPT3Predictions["rawPrediction"] = untunedGPT3Predictions.apply(lambda x: openai.Completion.create(model="text-davinci-003", 
                                    prompt=x["prompt"], 
                                    temperature=0, max_tokens=10).choices[0].text, axis=1 )
```

```python
untunedGPT3Predictions["rawPrediction"] = untunedGPT3Predictions["rawPrediction"].apply(lambda x: x.strip() )
untunedGPT3Predictions
```

```python
# Fine-tuning data. Note: right now, GPT-3 can only be fine-tuned on 'completion' tasks, not 'insertion tasks', so this doesn't really work.
fineTuningData = prompts.head(2)
fineTuningData = (fineTuningData.assign(completion=["5 {{STOP}}","95 {{STOP}}"]))[["prompt","completion"]]
openai.File.create(
  file=fineTuningData.to_json(orient = "records", lines=True),
  purpose='fine-tune'
)
```
